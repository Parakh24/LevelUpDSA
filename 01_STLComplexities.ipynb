```python
%%writefile practice.cpp

#include<bits/stdc++.h>

using namespace std;

int main() {

int n;
cin >> n;

while(n--)  {

  int t,x;
  cin >> t >> x;
  multiset<long long> bags;

  for(int i=0; i<t; i++){
      long long candy_ct;
      cin >> candy_ct;

      bags.insert(candy_ct);
  }

  long long total_candies = 0;

  for(int i=0; i<x; i++) {

      auto last_it = (--bags.end());
      total_candies += (*last_it);
      bags.erase(last_it);
      bags.insert((*last_it/2));

  }

  cout << total_candies << endl;


}

  return 0;
}
```

```python
%%writefile nesting2.cpp

#include<bits/stdc++.h>
using namespace std;

int main() {

map<pair<string,string> , vector<int>> m;

int n;
cin >> n;

for(int i=0; i<n; i++){

  string fn, ln;
  int x;
  cin >> fn >> ln >> x;

for(int j = 0; j<x; ++j) {

   int ct,
   cin >> ct;
   m[make_pair(fn,ln)].push_back(ct);

}

}

for(auto &pr : m) {

   auto &full_name = pr.first;
   auto &list = pr.second;
   cout << full_name.first << " " << full_name.second << endl;
   cout << list.size() << endl;

   for(auto &element : list){
       cout << element << "  ";
   }

   cout << endl;

}

  return 0;
}
```
